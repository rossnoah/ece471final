```python
import torch
import torchvision
import matplotlib.pyplot as plt

# seed so random stuff isnt random
SEED = 100


dataset = torchvision.datasets.EuroSAT(root='./data', download=True)

classes = ('AnnualCrop', 'Forest', 'HerbaceousVegetation', 'Highway',
           'Industrial', 'Pasture', 'PermanentCrop', 'Residential', 'River', 'SeaLake')



```

```python
# display an image from each class


images = {}

for image, label in dataset:
    if label not in images:
        images[label] = image
        if len(images) == 10:
            break


if len(images) == 10:
    figure, axes = plt.subplots(2, 5, figsize=(15, 6))
    for i, label in enumerate(images):
        image = images[label]
        ax = axes[i // 5, i % 5]
        ax.imshow(image)
        ax.set_title(classes[label])
        ax.axis('off')
    plt.tight_layout()
    plt.show()

```
